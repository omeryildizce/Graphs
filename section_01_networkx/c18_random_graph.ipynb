```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
er = nx.erdos_renyi_graph(100, 0.15)
plt.figure(figsize=(16,9))
nx.draw(er, node_color="#A0CBE2", with_labels = True)
plt.show()
```

```python
ws = nx.watts_strogatz_graph(30, 3, 0.1)
plt.figure(figsize=(16,9))
nx.draw(ws, node_color="#A0CBE2", with_labels = True)
```

```python
rd = nx.random_lobster(100, 0.9, 0.9)
plt.figure(figsize=(16,9))
nx.draw(rd, node_color="green", with_labels = True)
```
