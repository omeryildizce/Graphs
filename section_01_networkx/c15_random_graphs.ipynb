```python
import networkx as nx
import matplotlib.pyplot as plt
```

Draw random graph called grid_2d_graph

```python
G = nx.grid_2d_graph(4,4) # 4x4
pos = nx.spring_layout(G, iterations=100)

#firstplot
plt.figure(figsize=(16,16))
plt.subplot(221)
nx.draw(G,pos, font_size=8, node_size=500, node_color="gold")

#2nd plot
plt.subplot(222)
nx.draw(G,pos, with_labels=0, node_size=0, node_color="r")

#3rd plot
plt.subplot(223)
nx.draw(G, pos,with_labels=0, node_size=1, node_color="g")

#4th plot
plt.subplot(224)

# convert undirected graph to directed
H = G.to_directed()
nx.draw(H, with_labels=0, node_size=20, node_color="b", pos=pos)
plt.savefig("grid.png")
plt.show()
```
