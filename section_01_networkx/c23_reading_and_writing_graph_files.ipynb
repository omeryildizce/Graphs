```python
import networkx as nx
import matplotlib.pyplot as plt
```

Writing

```python
# create a ranom graph
G = nx.path_graph(5)
plt.figure(figsize=(16,9))
nx.draw(G, with_labels=True, node_size = 1500)
```

```python
# write file to disk
nx.write_gpickle(G, "path.gpickle")
```

```python
import os
os.listdir()
```

Reading

```python
data = nx.read_gpickle("./path.gpickle")
plt.figure(figsize=(16,9))
nx.draw(data, with_labels=True, node_size=5000)
```

```python
# writing edge list to a text file

nx.write_edgelist(nx.path_graph(20), "edge.txt", data=True)
```

```python
os.listdir()
```

```python
# reading the edge list file
G = nx.read_edgelist("edge.txt")

plt.figure(figsize=(16,9))
nx.draw(G, with_labels=True, node_size=5000)
```

```python
# files
file = open("edge.txt", "rb")
# open
G = nx.read_edgelist(file)
plt.figure(figsize=(16,9))
nx.draw(G, with_labels=True, node_size=5000)
# read from it

# finally close the input stream
file.close()
```

```python
nx.info(G)
```

```python
hartford = nx.read_edgelist("edge.txt", create_using=nx.DiGraph(), nodetype=int)

N, E = hartford.order(), hartford.size()

avg_deg  = float(E) / N
print("Node",N)
print("Edge",E)
print("Average Degree: ", avg_deg)
```
