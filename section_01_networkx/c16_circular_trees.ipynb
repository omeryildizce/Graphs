```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
SIZE = 100
CT = nx.star_graph(SIZE)

pos = nx.spring_layout(CT, iterations=20)
colors = range(SIZE)

plt.figure(figsize=(16,9))
nx.draw(CT, pos, node_color="#A0CBE2", edge_color=colors, width=4, edge_cmap=plt.cm.Blues)
```
