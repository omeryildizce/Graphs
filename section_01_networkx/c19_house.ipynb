```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
house = nx.house_graph()

# set positions of nodes

pos = {0:(0,0),1:(1,0),2:(0,1),3:(1,1), 4:(0.5,2)}
plt.figure(figsize=(16,9))

nx.draw_networkx(house, pos, node_color="gold", node_size=2000, nodelist=[4])
nx.draw_networkx(house, pos, node_color="c", node_size=3000, nodelist=[0,1,2,3])
nx.draw_networkx_edges(house, pos, edge_color="g", alpha=0.5, width=15)
plt.axis("off")
plt.show()
```
