```python
import networkx as nx
from networkx.algorithms import bipartite
import matplotlib.pyplot as plt
```

```python
B = nx.Graph()
# define cities
cities = ["Beijing", "Xian", "Vancouver", "San Francisco", "Austin", "Boston", "Istanbul"]

# define people
people = ["Theo", "Student" ]


# adding edges of cities and people
B.add_nodes_from(cities, bipartite='cities')
B.add_nodes_from(people, bipartite="people") 

#define edges
edges = [("Theo","Vancouver"), ("Student", "Xian"), ("Theo","San Francisco"), ("Student", "Boston") , ("Theo", "Boston"), ("Student", "Beijing"), ("Theo", "Istanbul")]

B.add_edges_from(edges)

plt.figure(figsize=(16,9))
nx.draw(B, with_labels=True, node_color="gold", node_size =750)
```
