```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
barbell = nx.barbell_graph(10, 10)
plt.figure(figsize=(16,9))
nx.draw(barbell, node_color = "gold", node_size=999, with_labels=True) 
```

```python
# 
loli = nx.lollipop_graph(10, 10) 
plt.figure(figsize=(16,9))
nx.draw(loli, node_color = "gold", node_size=999, with_labels=True) 
```

```python
# 
binom = nx.binomial_graph(100, 0.15) 
plt.figure(figsize=(16,9))
nx.draw(binom, node_color = "gold", node_size=999, with_labels=True) 
```

```python
chordal = nx.chordal_cycle_graph(20)
plt.figure(figsize=(16,9))
nx.draw(chordal, node_color = "gold", node_size=999, with_labels=True) 
```

```python
k_3_5 = nx.complete_bipartite_graph(3,5)

plt.figure(figsize=(16,9))
nx.draw(k_3_5, node_color = "gold", node_size=999, with_labels=True, edge_color="gold") 
```
