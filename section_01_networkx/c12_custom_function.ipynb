```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# define our function called simple-graph
def simple_graph():
    #create_our_empty graph
    G = nx.Graph()

    #add edges
    G.add_edge("A", "B")
    G.add_edge("B", "C")
    G.add_edge("C", "A")

    G.add_edge("A", "E")
    G.add_edge("E", "F")
    G.add_edge("F", "G")
    G.add_edge("G", "E")
    
    G.add_edge("B", "J")
    G.add_edge("J", "K")
    G.add_edge("K", "L")
    G.add_edge("L", "J")

    G.add_edge("C", "M")
    G.add_edge("M", "N")
    G.add_edge("N", "O")
    G.add_edge("O", "P")
    G.add_edge("P", "M")
    G.add_edge("P", "L")

    # draw graph
    plt.figure(figsize=(16,9))
    nx.draw(G, with_labels=True, node_color="gold", node_size=800)


```

```python
simple_graph()
```

how to delete nodes from a graph

```python
def delete_nodes_in_graph():
 
    G = nx.Graph()

    #add edges
    G.add_edge("A", "B")
    G.add_edge("B", "C")
    G.add_edge("C", "A")

    G.add_edge("A", "E")
    G.add_edge("E", "F")
    G.add_edge("F", "G")
    G.add_edge("G", "E")
    
    G.add_edge("B", "J")
    G.add_edge("J", "K")
    G.add_edge("K", "L")
    G.add_edge("L", "J")

    G.add_edge("C", "M")
    G.add_edge("M", "N")
    G.add_edge("N", "O")
    G.add_edge("O", "P")
    G.add_edge("P", "M")
    G.add_edge("P", "L")

    # draw graph
    plt.figure(figsize=(16,9))
    nx.draw(G, with_labels=True, node_color="gold", node_size=800)
    plt.title("Original Graph")
    plt.show()

    # remove node B from the graph
    G.remove_node("B")
    
    # draw graph
    plt.figure(figsize=(16,9))
    nx.draw(G, with_labels=True, node_color="gold", node_size=800)
    plt.title("Remove B")
    plt.show()

delete_nodes_in_graph()
```

customize node size and node colors in the graph

```python
def customize_node_size_and_node_colors():
 
    G = nx.Graph()

    #add edges
    G.add_edge("A", "B")
    G.add_edge("B", "C")
    G.add_edge("C", "A")
    G.add_edge("D", "A")

    G.add_edge("A", "E")
    G.add_edge("E", "F")
    G.add_edge("F", "G")
    G.add_edge("G", "E")
    
    G.add_edge("B", "J")
    G.add_edge("J", "K")
    G.add_edge("K", "L")
    G.add_edge("L", "J")

    G.add_edge("C", "M")
    G.add_edge("M", "N")
    G.add_edge("N", "O")
    G.add_edge("O", "P")
    G.add_edge("P", "M")
    G.add_edge("P", "L")
    
    G.add_edge("H", "D")
    G.add_edge("H", "I")
    G.add_edge("R", "D")
    G.add_edge("R", "I")
    G.add_edge("D", "I")




    # draw graph
    plt.figure(figsize=(16,9))
    nx.draw(G, with_labels=True, node_color="gold", node_size=800)
    plt.title("Original Graph")
    plt.show()

    # define the node list
    nodelst = ["A","B","C","D","E","F","G","H","I","J","K","L","M","N","O","P","R"]
    
    # define the color list
    colorlst = ["gold","r","g","b", "c","y","gold","r","g","b", "c","y","gold","r","g","b", "c"]
    # define the node size list
    nodesize = [1000,2000,3000,1000,2000,3000,1000,2000,3000,1000,2000,3000,1000,2000,3000,1000,2000]

    # draw graph
    plt.figure(figsize=(16,9))
    nx.draw(G, with_labels=True, node_color=colorlst, node_size=nodesize, nodelist=nodelst)
    plt.title("Graph")
    plt.show()
customize_node_size_and_node_colors()
```

costimize edge colors in the graph

```python
def customize_edge_colors():
 
    G = nx.Graph()

    #add edges
    G.add_edge("A", "B")
    G.add_edge("B", "C")
    G.add_edge("C", "A")


    # draw graph
    plt.figure(figsize=(16,9))
    nx.draw(G, with_labels=True, node_color="gold", node_size=800)
    plt.title("Original Graph")
    plt.show()

    # define the node list
    nodelst = ["A","B","C" ]
    
    # define the color list
    colorlst = ["b","r","g"]
    # define the node size list
    nodesize = [1000,2000,3000]

    # draw graph
    plt.figure(figsize=(16,9))
    nx.draw(G, with_labels=True, node_color="gold", edge_color=colorlst, node_size=nodesize, nodelist=nodelst)
    plt.title("Graph")
    plt.show()
customize_edge_colors()
```
