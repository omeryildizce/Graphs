```python
import networkx as nx
import matplotlib.pyplot as plt
```

shortest path from one node to another

```python
def pltfig():
    plt.figure(figsize=(16,9))

pltfig()
N = nx.path_graph(5, create_using=nx.DiGraph())
nx.draw_networkx(N, node_size=1000, node_color="gold")
```

find the shortest path from node 0 to node 4

```python
nx.shortest_path(N, source=0, target=4, method="dijkstra")
```

given only the source=0, we have no target specifed

```python
nx.shortest_path(N, source=0)
```

```python
p = nx.shortest_path(N, source=0)
p[3]
```

given the target = 4, with no source 

```python
p = nx.shortest_path(N, target=4)
p[3]
```

given no target and no source

```python
p = nx.shortest_path(N)
p
```
