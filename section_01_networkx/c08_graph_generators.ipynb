```python
import networkx as nx
import matplotlib.pyplot as plt
```

draw a complete-graph

```python
plt.figure(figsize=(16,9))
cg = nx.complete_graph(5, create_using=nx.DiGraph())
nx.draw_networkx(cg)
```

```python
cg.in_degree()
```

```python
cg.out_degree()
```

draw a graph called path-graph

```python
plt.figure(figsize=(16,9))
pg = nx.path_graph(5, create_using=nx.DiGraph())
nx.draw_networkx(pg, node_size = 500)
plt.show()
```

draw a gnr-graph, which is like a facebook or twitter

```python
plt.figure(figsize=(16,9))
gnr = nx.gnr_graph(100,1, create_using=nx.DiGraph())
nx.draw_networkx(gnr)
plt.show()
```

```python
gnr.degree
```

draw complete-bipartite-graph

```python
plt.figure(figsize=(16,9))
cbg = nx.complete_bipartite_graph(50,10)
nx.draw_networkx(cbg)
```

```python
cbg.order()
```

```python
cbg.size()
```

draw grid_graph

```python
plt.figure(figsize=(16,9))
gg = nx.grid_graph(dim=[5,5] )
nx.draw_networkx(gg, node_size=1000, node_color="b", font_color="white")
```

```python
gg.order()
```

```python
gg.size()
```

```python
gg.is_directed()
```
