Facebook Analysis

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# read in file
fb = nx.read_edgelist("facebook_combined.txt", create_using=nx.Graph(), nodetype=int)
```

```python
# create a spring-layout of fb
pos = nx.spring_layout(fb)
plt.figure(figsize=(16,9))
plt.axis("off")
nx.draw_networkx(fb, pos=pos, with_labels = False, node_size = 350, node_color = "gold")
```

```python
nx.betweenness_centrality(fb, normalized=True)
```

```python
fb.degree()
```

```python
fb.neighbors(107)
```

```python
fb.order()
```

```python
fb.size()
```

```python
nx.triangles(fb)
```

```python
nx.clustering(fb)
```

```python
# draw histogram of degrees
plt.figure(figsize=(16,9))
hst = nx.degree_histogram(fb)
plt.hist(hst, bins=10, color="gold")
```

```python
# draw a bar chart of the betweenness centrality
bc = nx.betweenness_centrality(fb)
plt.figure(figsize=(16,9))
plt.bar(range(len(bc)), bc.values(), align="center")
```
