```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
k = nx.karate_club_graph()
plt.figure(figsize=(16,9))
nx.draw(k, node_color="gold", node_size = 800, with_labels = True)
```

```python
k.is_directed()
```

```python
k.name, nx.number_of_nodes(k), nx.number_of_edges(k), nx.average_node_connectivity(k)
```
