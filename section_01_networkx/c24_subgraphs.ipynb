```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
er = nx.erdos_renyi_graph(100, 0.15)
plt.figure(figsize=(16,9))
nx.draw(er, with_labels = True, node_size = 800)
```

```python
sub = nx.subgraph(er, [54,35,45,76,59,90,4,99])
plt.figure(figsize=(16,9))
nx.draw(sub, with_labels = True, node_size = 800)
```
