```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
p = nx.petersen_graph()

plt.figure(figsize=(16,9),)
nx.draw(p, node_color="cyan", node_size = 650, with_labels=True)
```

```python
t = nx.tutte_graph()
plt.figure(figsize=(16,9))
nx.draw(t, node_color="gold", node_size=700, with_labels=True)
```

```python
s = nx.sedgewick_maze_graph()
plt.figure(figsize=(16,9))
nx.draw(s, node_color="b", node_size=1700, with_labels=True, font_color="w")
```

```python
tt = nx.tetrahedral_graph()
plt.figure(figsize=(16,9))
nx.draw(tt, node_color="k", node_size=1700, with_labels=True, font_color="w")
```
